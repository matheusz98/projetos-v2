```python
import pandas as pd
```

### Juntando os dois datasets.

```python
# dataframe 1
df1 = pd.read_csv("gasolina_2000.csv", index_col=0)

df1
```

```python
# dataframe 2
df2 = pd.read_csv("gasolina_2010.csv", index_col=0)

df2
```

```python
# juntando ambos

df = pd.concat([df1, df2])

df
```

### Converta o DATA INICIAL e DATA FINAL para datetime.

```python
df["DATA INICIAL"] = pd.to_datetime(df["DATA INICIAL"])
df["DATA FINAL"] = pd.to_datetime(df["DATA FINAL"])
```

```python
type(df["DATA INICIAL"].iloc[2])
```

```python
df["DATA INICIAL"][2]
```

### Crie uma nova coluna para apresentar o ano e o mês (aaaa-mm) utilizando a coluna DATA FINAL como referência.

```python
df["ANO-MES"] = df["DATA FINAL"].apply(lambda x: '{}'.format(x.year)) + df["DATA FINAL"].apply(lambda x: '--{:02d}'.format(x.month))
```

```python
df["ANO-MES"]
```

### Liste todos os tipos de produtos na base de dados.

### Filtre o DataFrame para obter apenas dados da 'GASOLINA COMUM'. Grave em uma nova variável.

```python
df_filtro = df["PRODUTO"] == "GASOLINA COMUM"
```

```python
df2 = df[df_filtro]
```

```python
df2
```

### Qual o preço médio da revenda da gasolina em agosto de 2008?

```python
df2[df2['ANO-MES'] == '2008--08']["PREÇO MÉDIO REVENDA"].mean()
```

### Qual o preço médio de revenda da gasolina em maio de 2014 em São Paulo?

```python
df2[(df2["ANO-MES"] == '2014--05') & (df2['ESTADO'] == 'SAO PAULO')]["PREÇO MÉDIO REVENDA"].mean()
```

### Você conseguiria descobrir em qual(quais) estado(s) a gasolina ultrapassou a barreira dos R$ 5,00? E quando isso ocorreu?

```python
df2[df2["PREÇO MÉDIO REVENDA"] > 5][["ESTADO", "ANO-MES", "PREÇO MÉDIO REVENDA"]]
```

### Qual a média de preço dos estados da região sul em 2012?

```python
df_aux = df2[df2["DATA FINAL"].apply(lambda x: x.year) == 2012]

df_aux[df_aux["REGIÃO"] == "SUL"]["PREÇO MÉDIO REVENDA"].mean()
```

### Você conseguiria obter uma tabela contendo a variação percentual ano a ano para o estado do Rio de Janeiro?

```python
df2["MES"] = df2["DATA FINAL"].apply(lambda x: x.month)
df_rio = df2[df2["ESTADO"] == "RIO DE JANEIRO"]

df_month_rio = df_rio.groupby("ANO-MES")[["PREÇO MÉDIO REVENDA", "MES"]].last()

df_month_rio[df_month_rio["MES"] == 12]
```
